```python
import pandas as pd
```

```python
df = pd.read_csv("Loan_default.csv")
```

```python
df = df.drop(columns=["LoanID"])
```

```python
df.head(4)
```

```python
df["Loan_to_Income"] = df["LoanAmount"] / df["Income"]
```

```python
df
```

```python
df["Estimated_EMI"] = (df["LoanAmount"] * df["InterestRate"]) / (12 * 100)
```

```python
df
```

```python
df["EMI_to_Income"] = df["Estimated_EMI"] / df["Income"]
```

```python
df["High_DTI_Flag"] = (df["DTIRatio"] > 0.4).astype(int)
```

```python
df["Low_Credit_Flag"] = (df["CreditScore"] < 600).astype(int)
```

```python
df["Low_Employment_Stability"] = (df["MonthsEmployed"] < 12).astype(int)
```

```python
df["CreditLines_per_Age"] = df["NumCreditLines"] / df["Age"]
```

```python
df["Thin_Credit_File"] = (df["NumCreditLines"] <= 1).astype(int)
```

```python
df["Stable_Employment"] = df["EmploymentType"].isin(
    ["Government", "Salaried"]
).astype(int)
```

```python
df["Risk_Buffer"] = (
    (df["HasCoSigner"] == "Yes") |
    (df["MaritalStatus"] == "Married")
).astype(int)
```

```python
df['Education'].value_counts()
```

```python
education_map = {
    "High School": 0,
    "Bachelor's": 1,
    "Master's": 2,
    "PhD": 3
}

df["Education_Level"] = df["Education"].map(education_map)
```

```python
df.groupby("Default")[[
    'Loan_to_Income',
    'EMI_to_Income',
    'High_DTI_Flag',
    'Low_Credit_Flag'
]].mean()
```

```python
df.dtypes
```

```python
x = df.drop(columns=["Default"])
y = df["Default"]
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_idx, test_idx in split.split(x,y):
    x_train = x.iloc[train_idx]
    x_test = x.iloc[test_idx]

    y_train = y.iloc[train_idx]
    y_test = y.iloc[test_idx]
```

```python
x_train.isna().sum()
```

```python
num_attribs = [
    "Age",
    "Income",
    "LoanAmount",
    "CreditScore",
    "MonthsEmployed",
    "NumCreditLines",
    "InterestRate",
    "LoanTerm",
    "DTIRatio",

    # Engineered features
    "Loan_to_Income",
    "Estimated_EMI",
    "EMI_to_Income",
    "High_DTI_Flag",
    "Low_Credit_Flag",
    "Low_Employment_Stability",
    "CreditLines_per_Age",
    "Thin_Credit_File",
    "Stable_Employment",
    "Risk_Buffer",
    "Education_Level"
]


cat_attribs = [
    "Education",
    "EmploymentType",
    "MaritalStatus",
    "HasMortgage",
    "HasDependents",
    "LoanPurpose",
    "HasCoSigner"
]
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline(steps=[
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore", drop="first"))
])
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs)
])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    roc_auc_score,
    classification_report,
    confusion_matrix
)

# Build pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(
        class_weight='balanced',
        max_iter=1000,
        random_state=42
    ))
])

# Train
model.fit(x_train, y_train)

# Predict probabilities
y_proba = model.predict_proba(x_test)[:, 1]

# Apply custom threshold
threshold = 0.40
y_pred = (y_proba >= threshold).astype(int)

# Evaluate
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print(f"Confusion Matrix (threshold = {threshold}):")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    roc_auc_score,
    classification_report,
    confusion_matrix
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(
        class_weight='balanced',
        max_depth=6,
        min_samples_leaf=50,
        random_state=42
    ))
])

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_proba = model.predict_proba(x_test)[:, 1]
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    roc_auc_score,
    classification_report,
    confusion_matrix
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        class_weight='balanced',
        n_estimators=200,
        max_depth=12,
        min_samples_leaf=50,
        random_state=42,
        n_jobs=-1
    ))
])

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_proba = model.predict_proba(x_test)[:, 1]
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
